```python
!pip install -U datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import matplotlib.pyplot as plt
```

## ðŸŸ© Bar Chart with Labels (1.30.1)

```python
values = df['job_title_short'].unique()
counts = df['job_title_short'].value_counts()
counts.plot(kind='bar')
plt.ylabel("NUmber of jobs per job title")
plt.xlabel("Number of jobs titles")
plt.xticks(rotation=45,ha='right')
plt.title("Jobs available in each category")
```

---

## ðŸŸ¨ Horizontal Bar Chart with Labels (1.30.2)

```python
values = df['job_title_short'].unique()
counts = df['job_title_short'].value_counts().sort_values()
counts.plot(kind='barh')
plt.ylabel("NUmber of jobs per job title")
plt.xlabel("Number of jobs titles")
plt.xticks(rotation=45,ha='right')
plt.title("Jobs available in each category")
```

---

## ðŸŸ¥ Line Chart with Labels (1.30.3)

```python
values = df['job_title_short'].unique()
counts = df['job_title_short'].value_counts().sort_values()
counts.plot(kind='line')
plt.ylabel("NUmber of jobs per job title")
plt.xlabel("Number of jobs titles")
plt.xticks(ticks=range(len(counts)),labels = counts.index,rotation=45,ha='right')
plt.title("Jobs available in each category")
plt.tight_layout()
plt.figure(figsize=(12,6))
```
