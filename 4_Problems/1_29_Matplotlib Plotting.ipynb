```python
!pip install -U datasets
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import matplotlib.pyplot as plt
```

## ðŸŸ© Job Country Bar Chart (1.29.1)

```python
country_counts = df['job_title_short'].value_counts()
country_counts.plot(kind = 'bar')
plt.figure(figsize=(1000,200))
```

---

## ðŸŸ¨ Average Salary Horizontal Bar Chart (1.29.2)

```python
salary = df.groupby('job_title_short')['salary_year_avg'].mean()
salary.plot(kind='bar')
```

---

## ðŸŸ¥ Average Salary Trend Line Plot (1.29.3)

```python
salaries = df.groupby('job_title_short')[['salary_year_avg','salary_hour_avg']].mean()
plt.figure(figsize=(20,16))
salaries.plot()
job_titles = salaries.index
tick_positions = range(len(job_titles))
#tick_positions
plt.xticks(tick_positions,job_titles,rotation = 45,ha='right')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
```
