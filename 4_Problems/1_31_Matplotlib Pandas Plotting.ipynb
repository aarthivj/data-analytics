```python
!pip install -U datasets
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import matplotlib.pyplot as plt
```

## 🟩 Job Posting Location Bar Plot (1.31.1)

```python
values_filtered = df['job_title_short'].value_counts().sort_values(ascending=False)
values_filtered
```

```python
df_filtered = df[df['job_title_short'].isin(['Data Analyst','Data Engineer','Data Scientist'])]
df_filtered
```

```python
df_filtered.groupby('job_location')['job_title_short'].value_counts()
```

```python
pivot_df = df_filtered.pivot_table(index='job_location',columns='job_title_short',aggfunc='size')
pivot_df.dropna()
```

```python
pivot_df.plot(kind='bar',figsize=(25,8))
```

```python
df['job_posted_month'].value_counts()
```

```python
df['job_title_short'].value_counts()
```
