## ðŸŸ¨ Days Since Posted (2.10.2)

```python
from datetime import datetime
today = pd.to_datetime(datetime.today().date())
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['days_since_posted'] = (today - df['job_posted_date']).dt.days
display(df[['job_posted_date', 'days_since_posted']].head())
```

```python
from datetime import datetime
import pandas as pd

# Method to calculate days between today and a given date
def days_since_posted(posted_date):
    today = pd.to_datetime(datetime.today().date())
    posted_date = pd.to_datetime(posted_date)
    return (today - posted_date).days

# Apply this method to the column
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['days_since_posted'] = df['job_posted_date'].apply(days_since_posted)

# Optional: Display sample
display(df[['job_posted_date', 'days_since_posted']].head())
```

## ðŸŸ¥ Salary Category (2.10.3)

```python
df_altered = df.dropna(subset = ['salary_year_avg']).copy()
display(df_altered.head())
```

---

```python
def sal_category(salary):
  if salary > 10000:
    return "Low"
  elif salary > 20000:
    return "Medium"
  else:
    return "High"
```

```python
 df_altered['salary_category'] =  df_altered['salary_year_avg'].apply(sal_category)
```
